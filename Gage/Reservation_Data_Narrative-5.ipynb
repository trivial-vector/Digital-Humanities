# Cleaning the Age Column

```python
import pandas as pd
import numpy as np
import xlrd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

# Reading in Census Data

```python
xls = pd.ExcelFile('data/Reservation Census.xlsx')
df_1900 = pd.read_excel(xls, '1900 Census')
df_1910 = pd.read_excel(xls, '1910 Census')
df_1920 = pd.read_excel(xls, '1920 Census')
```

## Preview of 1900 - 1920 Census Data

```python
df_1900['Year'] = 1900
df_1910['Year'] = 1910
df_1920['Year'] = 1920
df_1900_rsac = df_1900[["Age at last birthday (8)",'Year']]
df_1910_rsac = df_1910[['Age at last birthday (8)','Year']]
df_1920_rsac = df_1920[["Age at last birthday (8)",'Year']]
```

```python
df_1900_rsac.rename(columns={'Age at last birthday (8)': 'Age'}, inplace=True)
df_1900_rsac.head()
```

## Preview of 1910 Census Data

```python
df_1910_rsac.rename(columns={'Age at last birthday (8)': 'Age'}, inplace=True)
df_1910_rsac.head()
```

## Preview of 1920 Census Data

```python
df_1920_rsac.rename(columns={'Age at last birthday (8)': 'Age'}, inplace=True)
df_1920_rsac.head()
```

```python
# converts a column into string objects
df_1900_age = df_1900_rsac["Age"].astype('str')

for i in range (0 ,len(df_1900_age)):    
    num_list = [int(s) for s in re.findall(r'\b\d+\b', str(df_1900_age[i])  )]

    if len(num_list) == 1:
        df_1900_age[i] = round(num_list[0],1)
    elif len(num_list) == 2:
        df_1900_age[i] = round((num_list[0] + (num_list[1]/12)),1)    
    else:
        df_1900_age[i] = df_1900_rsac["Age"][i]   

# Substitute the cleaned values values back into the original dataframe
df_1900_rsac["Age"] = df_1900_age.values

# View the cleaned dataframe
df_1900_rsac.head(50)
```
